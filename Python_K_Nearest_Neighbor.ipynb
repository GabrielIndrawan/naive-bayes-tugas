```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("online_gaming_behavior_dataset.csv")
data.head()
```

```python
data = pd.get_dummies(data, columns=['Gender','Location','GameGenre','GameDifficulty'])
data.head()
```

```python

x = data.iloc[:, data.columns != "EngagementLevel"].values
y = data["EngagementLevel"].values
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
train_x , test_x, train_y, test_y = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from os import preadv
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
```

```python
test_x = scaler.fit_transform(test_x)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=7, metric="minkowski", p=2)
model.fit(train_x,train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y,y_pred)
print(cm)
```
